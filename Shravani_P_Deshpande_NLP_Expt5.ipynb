```python
documents = [
    "Hi my name is Shravani",
    "I study BTech at MIT WPU Pune",
    "NLP is a nice subject"
]

```

```python
import string
import math
```

```python
def preprocess(text):
    text = text.lower()
    text = text.translate(str.maketrans('', '', string.punctuation))
    tokens = text.split()
    return tokens

processed_docs = [preprocess(doc) for doc in documents]

print("Preprocessed Documents:")
for i, doc in enumerate(processed_docs):
    print(f"D{i+1}:", doc)
```

```python
vocabulary = sorted(set(word for doc in processed_docs for word in doc))
print("\nVocabulary:")
print(vocabulary)
```

```python
bow_matrix = []

for doc in processed_docs:
    row = []
    for word in vocabulary:
        row.append(doc.count(word))
    bow_matrix.append(row)

print("\nBag of Words Matrix:")
print("Words:", vocabulary)
for i, row in enumerate(bow_matrix):
    print(f"D{i+1}:", row)
```

```python
tf_matrix = []

for i, doc in enumerate(processed_docs):
    tf_row = []
    total_words = len(doc)
    for count in bow_matrix[i]:
        tf_row.append(count / total_words)
    tf_matrix.append(tf_row)

print("\nTerm Frequency (TF) Matrix:")
for i, row in enumerate(tf_matrix):
    print(f"D{i+1}:", row)
```

```python
N = len(processed_docs)
idf = []

for i, word in enumerate(vocabulary):
    df = sum(1 for doc in processed_docs if word in doc)
    idf.append(math.log(N / df))

print("\nInverse Document Frequency (IDF):")
for word, value in zip(vocabulary, idf):
    print(f"{word}: {value:.4f}")
```

```python
tfidf_matrix = []

for tf_row in tf_matrix:
    tfidf_row = []
    for tf, idf_value in zip(tf_row, idf):
        tfidf_row.append(tf * idf_value)
    tfidf_matrix.append(tfidf_row)

print("\nTF-IDF Matrix:")
for i, row in enumerate(tfidf_matrix):
    print(f"D{i+1}:", row)
```

```python
print("\n--- Final Results ---")
print("Vocabulary:", vocabulary)

print("\nBag of Words Matrix:")
for i, row in enumerate(bow_matrix):
    print(f"D{i+1}:", row)

print("\nTF-IDF Matrix:")
for i, row in enumerate(tfidf_matrix):
    print(f"D{i+1}:", row)
```
